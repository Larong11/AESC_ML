##### Задача 1. Построение "среднего" алгоритма
В этом задании вам нужно построить графики, демонстрирующие, как алгоритм аппроксимирует истинную зависимость в данных и как он меняется в зависимости от гиперпараметров метода обучения.

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def f(x):
    return np.sin(x)  # истинная зависимость в данных
sample_size = 100      # длина выборки
samples_num = 20       # количество выборок
linspace = np.linspace(0, 7, 1000)  # точки для построения графиков
```

__1. (2 балла)__
1. Сгенерируйте выборку $x$ из одномерного экспоненциального распределения (np.random.exponential) длины sample_size.
1. Создайте вектор целевых переменных $y$ как сумму $f(x)$ и случайного шума, сгенерированного из равномерного распределения на отрезке $[-1, 1]$ (np.random.uniform).
1. Обучите DecisionTreeRegressor с параметрами по умолчанию на полученной выборке и сделайте предсказания для объектов из linspace.
1. Постройте два графика на одном рисунке: $f(x)$ и зависимость, восстановленную решающим деревом.

    Рекомендация: не забудьте, что все методы обучения в sklearn требуют на вход двумерную матрицу объекты-признаки. Сделать такую из одномерного вектора можно добавлением мнимых осей (np.newaxis).

```python
from sklearn.tree import DecisionTreeRegressor

x = np.random.exponential(scale=1.0, size=sample_size)
noise = np.random.uniform(-1, 1, size=sample_size)
y = f(x) + noise

regressor = DecisionTreeRegressor()
regressor.fit(x[:, np.newaxis], y)

predictions = regressor.predict(linspace[:, np.newaxis])

plt.figure(figsize=(10, 6))

plt.plot(linspace, f(linspace), label='$f(x) = sin(x)$', color='blue')
plt.plot(linspace, predictions, label='Decision Tree Prediction', color='red')

plt.scatter(x, y, color='green', alpha=0.5, label='Training samples')
plt.title('Decision Tree Regression vs True Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

__2. (1 балл)__

Повторите первые 3 шага, описанные выше, samples_num раз. На одном графике для каждого обученного решающего дерева визуализируйте восстановленную им зависимость (рекомендуется все такие линии рисовать полупрозрачными и серым цветом: plt.plot(...... color="gray", alpha=0.5)).  На этом же графике изобразите истинную зависимость f(x) (красным цветом: color="red") и усредненную по всем деревьям восстановленную зависимость (черным цветом: color="black").    

```python
def depth(a):
    predictions_all = []

    plt.figure(figsize=(12, 8))

    for _ in range(samples_num):
        x = np.random.exponential(size=sample_size)

        noise = np.random.uniform(-1, 1, size=sample_size)
        y = f(x) + noise

        # 3. Обучение DecisionTreeRegressor
        regressor = DecisionTreeRegressor(max_depth=a)
        regressor.fit(x[:, np.newaxis], y)

        # Предсказания на линспейсе
        predictions = regressor.predict(linspace[:, np.newaxis])
        predictions_all.append(predictions)

        # Рисуем отдельное дерево серой полупрозрачной линией
        plt.plot(linspace, predictions, color="gray", alpha=0.5)

    # Истинная функция
    plt.plot(linspace, f(linspace), color="red", label='True function $f(x) = sin(x)$')

    # Усредненная предсказанная зависимость
    mean_predictions = np.mean(predictions_all, axis=0)
    plt.plot(linspace, mean_predictions, color="black", label='Mean prediction')

    plt.title('Decision Trees Predictions vs True Function')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

depth(None)
```

```python
import pandas as pd
data = pd.read_csv('students.csv')
data.head()
```

```python
from importlib import reload
from matplotlib import pyplot as plt
import hw3code
reload(hw3code)
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
from sklearn.metrics import roc_auc_score
# Функция для расчета критерия Джини при заданном пороге
def gini_at_threshold(feature, threshold, y):
    y_pred = (feature > threshold).astype(int)
    auc = roc_auc_score(y, y_pred)
    gini = 2 * auc - 1
    return gini

# Построим кривые "порог — значение критерия Джини" для всех признаков
plt.figure(figsize=(14, 8))
for col in X.columns:
    thresholds = np.linspace(X[col].min(), X[col].max(), 100)
    ginis = [gini_at_threshold(X[col], t, y) for t in thresholds]
    plt.plot(thresholds, ginis, label=col)

plt.xlabel('Threshold')
plt.ylabel('Gini')
plt.title('Gini vs Threshold for each feature')
plt.legend()
plt.grid(True)
plt.show()

# Построим scatter-графики "значение признака — класс"
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for i, col in enumerate(X.columns):
    axes[i].scatter(X[col], y, alpha=0.5)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Class')
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```
